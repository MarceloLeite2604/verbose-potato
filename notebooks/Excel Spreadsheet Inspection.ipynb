# Excel Spreadsheet Inpection

## Spreadsheet Cells Information Gathering

```python
import openpyxl
from openpyxl.styles.colors import RGB 
import json

spreadsheet_location = 'input-files/ferramenta_ghg_protocol_v2025.0.1.xlsx'

def extract_spreadsheet_data(file_path):
    wb = openpyxl.load_workbook(file_path, data_only=False)  # Load formulas
    data = {}

    sheetnames = wb.sheetnames[4:5]

    for sheet in sheetnames:
        ws = wb[sheet]
        sheet_data = []

        for row in ws.iter_rows():
            for cell in row:
                cell_data = {
                    "cell": cell.coordinate,
                    "value": cell.value if cell.value else None,
                    # "formula": cell.formula if cell.data_type == 'f' and cell.formula else None,
                    "metadata": {}
                }

                # Extract comments
                if cell.comment:
                    cell_data["metadata"]["comment"] = cell.comment.text

                if cell.fill and cell.fill.bgColor and cell.fill.bgColor and cell.fill.bgColor.rgb and cell.fill.bgColor.rgb != '00000000':

                    print(f'Cell: {cell.coordinate} background color: {cell.fill.bgColor.rgb}.')

                sheet_data.append(cell_data)
        
        data[sheet] = sheet_data

    return data

# Example usage
structured_data = extract_spreadsheet_data(spreadsheet_location)

# Save structured data to a JSON file for further processing
# with open("spreadsheet_data.json", "w") as f:
#     json.dump(structured_data, f, indent=4)

# print(json.dumps(structured_data, indent=2))
```
